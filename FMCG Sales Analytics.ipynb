**FMCG Sales Analytics Project**

**Creating sample dataset for sales via pandas**

```python
import pandas as pd

data = {
    "date": ["2024-01-01","2024-01-02","2024-01-03","2024-01-04","2024-01-05"],
    "region": ["North","South","East","West","North"],
    "distributor": ["D1","D2","D3","D1","D2"],
    "product": ["Soap","Shampoo","Soap","Toothpaste","Shampoo"],
    "sales_amount": [1200,1500,900,1100,1700],
    "inventory": [200,150,180,100,160]
}

df = pd.DataFrame(data)

df
```

**Importing dataset frame into csv file**

```python
df.to_csv("fmcg_sales_data.csv", index=False)
```

# New section

```python
total_sales = df["sales_amount"].sum()
print("Total Sales:", total_sales)
```

```python
sales_by_region = df.groupby("region")["sales_amount"].sum()
print(sales_by_region)
```

```python
sales_by_product = df.groupby("product")["sales_amount"].sum()
print(sales_by_product)
```

```python
import matplotlib.pyplot as plt

sales_by_region.plot(kind="bar")

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales Amount")

plt.show()
```

```python
sales_by_product.plot(kind="bar")

plt.title("Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales Amount")

plt.show()
```

```python
df["date"] = pd.to_datetime(df["date"])

sales_trend = df.groupby("date")["sales_amount"].sum()

sales_trend.plot()

import matplotlib.pyplot as plt
plt.title("Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales Amount")

plt.show()
```

```python
sales_by_distributor = df.groupby("distributor")["sales_amount"].sum()

sales_by_distributor.plot(kind="bar")

import matplotlib.pyplot as plt
plt.title("Sales by Distributor")
plt.xlabel("Distributor")
plt.ylabel("Sales Amount")

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

df["day_number"] = range(len(df))

X = df[["day_number"]]
y = df["sales_amount"]

model = LinearRegression()
model.fit(X, y)

future_day = np.array([[6]])

predicted_sales = model.predict(future_day)

print("Predicted Sales for Next Day:", predicted_sales[0])
```
